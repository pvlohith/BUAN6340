# Adding columns PA, Average and OBP
## PA(Plate Appearances) = AB+BB+IBB+SH+SF
## Average = H/AB
## OBP(on base percentage) = (H+BB+IBB+SH+SF)/AB

```python
df['PA']=df['AB']+df['BB']+df['IBB']+df['SH']+df['SF']
df['Average'] = df['H']/df['AB']
df['OBP'] = (df['H']+df['BB']+df['IBB']+df['SH']+df['SF'])/df['AB']
```
