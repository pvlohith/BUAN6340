```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
all_stocks=pd.read_csv('all_stocks_2006-01-01_to_2018-01-01.csv')
all_stocks['Date']=pd.to_datetime(all_stocks['Date'],infer_datetime_format=True) # changing the Date column to Date type
all_stocks.index = all_stocks['Date'] #Chaging the index column to Date making the data a time series data
```

### Scatter plot for all the stocks closing prices. Analysis from this plot is not easy because of the large variation.

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 12)
sns.stripplot(x='Date',y='Close',hue="Name",data=all_stocks)
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.title('Scatter plot for all the stocks')
```

```python
amazon=pd.read_csv('AMZN_2006-01-01_to_2018-01-01.csv')
amazon['Date']=pd.to_datetime(amazon['Date'],infer_datetime_format=True)
amazon.index = amazon['Date']
```

```python
google=pd.read_csv('GOOGL_2006-01-01_to_2018-01-01.csv')
google['Date']=pd.to_datetime(google['Date'],infer_datetime_format=True)
google.index = google['Date']
```

### Plot showing the "close" price trend over the years for the 6 chosen companies.

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 12)
plt.plot(apple['Date'],apple['Close'],color='black',label='apple')
plt.plot(amazon['Date'],amazon['Close'],color='red',label='amazon')
plt.plot(google['Date'],google['Close'],color='blue',label='google')
plt.plot(microsoft['Date'],microsoft['Close'],color='green',label='microsoft')
plt.plot(cisco['Date'],cisco['Close'],color='cyan',label='cisco')
plt.plot(intel['Date'],intel['Close'],color='orange',label='intel')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Close price')
plt.legend(loc='upper left')
plt.title('Plot showing the "close" price trend over the years for the 6 chosen companies')
plt.show()
```

### Boxplots showing the quartile distribution of 'close' prices from 2016-2017

### Plot of volume of shares traded over time

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 12)
plt.plot(apple['Date'],apple['Volume'],color='black',label='apple',alpha=0.3)
plt.plot(amazon['Date'],amazon['Volume'],color='red',label='amazon')
plt.plot(google['Date'],google['Volume'],color='blue',label='google')
plt.plot(microsoft['Date'],microsoft['Volume'],color='green',label='microsoft')
plt.plot(cisco['Date'],cisco['Volume'],color='cyan',label='cisco')
plt.plot(intel['Date'],intel['Volume'],color='orange',label='intel',alpha=0.8)
plt.xticks(rotation=45)
plt.ylabel('in billions')
plt.legend(loc='upper left')
plt.show()
```

### Plot showing variation of total values of shares traded each day across years

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 12)
plt.plot(apple['Date'],apple['Volume']*apple['Close'],color='black',label='apple',alpha=0.5)
plt.plot(amazon['Date'],amazon['Volume']*amazon['Close'],color='red',label='amazon')
plt.plot(google['Date'],google['Volume']*google['Close'],color='blue',label='google')
plt.plot(microsoft['Date'],microsoft['Volume']*microsoft['Close'],color='green',label='microsoft')
plt.plot(cisco['Date'],cisco['Volume']*cisco['Close'],color='cyan',label='cisco')
plt.plot(intel['Date'],intel['Volume']*intel['Close'],color='orange',label='intel')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.show()
```

### Plot showing the daily difference betweeen open and close price over years. sub-plot shows the day with the max change.

```python
apple['Change']=apple['Close']-apple['Open']
fig, ax = plt.subplots()
fig.set_size_inches(18, 12)
plt.subplot(2,1,1)
plt.plot(apple['Date'],apple['Change'])
plt.xticks(rotation=45)
plt.subplot(2,1,2)
view=apple['2015-08-20':'2015-08-27']
plt.plot(view['Date'],view['Change'])
plt.xticks(rotation=45)
```

### Plot showing the daily difference betweeen daily high and low price over years. sub-plot shows the day with the max change.
### This plot along with the above one show that max change in both high and low, and in open and close occured at same time.

```python
apple['Change']=apple['High']-apple['Low']
fig, ax = plt.subplots()
fig.set_size_inches(18, 12)
plt.subplot(2,1,1)
plt.plot(apple['Date'],apple['Change'])
plt.xticks(rotation=45)
plt.subplot(2,1,2)
view=apple['2015-08-20':'2015-08-27']
plt.plot(view['Date'],view['Change'])
plt.xticks(rotation=45)
```

### Plots showing the effect of release of iphone on share prices for 2016 and 2017
